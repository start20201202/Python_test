```python
#Створюємо БД boosters.db
import sqlite3
```

```python
con = sqlite3.connect('boosters.db')
curs = con.cursor()
```

```python
#Створюємо таблицю в БД
curs.execute('''
    create table purchases (
    acquisition_date date,
    order_date date,
    subscription_id text,
    order_id text,
    order_status text,
    product_name text,
    rebill_number int,
    refund_date date,
    dispute_date date, 
    cancel_date date
    )
''')
```

```python
#дані з файлу task_2_purchases.csv додаємо в таблицю purchases
for acquisition_date, order_date, subscription_id, order_id, order_status, product_name,rebill_number, refund_date, dispute_date,cancel_date in zip(df['acquisition_date'],df['order_date'],df['subscription_id'],df['order_id'],df['order_status'],df['product_name'],df['rebill_number'],df['refund_date'],df['dispute_date'],df['cancel_date']):
    curs.execute(f"insert into purchases values ('{acquisition_date}','{order_date}','{subscription_id}','{order_id}','{order_status}','{product_name}','{rebill_number}','{refund_date}','{dispute_date}','{cancel_date}')")
```

```python
#перевіряємо чи додались дані в таблицю purchases
curs.execute('select * from purchases as t')
curs.fetchall()
```

```python
#Створюємо таблицю products
curs.execute('''
    create table products (
    product_name text,
    trial_price float,
    product_price float,
    trial_duration float,
    rebill_duration int
    )
''')
```

```python
#дані з файлу task_2_products.csv додаємо в таблицю products
for product_name,trial_price,product_price,trial_duration,rebill_duration in zip(df2['product_name'],df2['trial_price'],df2['product_price'],df2['trial_duration'],df2['rebill_duration']):
    curs.execute(f"insert into products values ('{product_name}','{trial_price}','{product_price}','{trial_duration}','{rebill_duration}')")
```

```python
#перевіряємо чи додались дані в таблицю products
result = pd.read_sql_query('select * from products as t limit 4', con)
result
```

```python
#рахуємо показники
#1.p2p (purch to pay) – який відсоток людей перейшли з пробного періоду з повноцінну оплату підписки;
#2.rebill rate 6m – середня довжина ребільного циклу на юзерів, які успішно пройшли тріал, за 6 місяців;
#3.refund rate – відсоток користувачів, які отримали рефанд, від тих хто взяв підписку;
#4.dispute rate – відсоток користувачів, які відкрили dispute, від тих хто взяв підписку.
#5.cancel rate – відсоток користувачів, яка закенселили підписку.
result = pd.read_sql_query( 
'select product_name, round(count(distinct case when rebill_number=1 then t.subscription_id end)\
*1.0/count(distinct t.subscription_id), 2) as "%_subscription_customers",\
avg(case when t.rebill_number>=1 and ((julianday(t.order_date)-julianday(t.acquisition_date))/30)<=6 \
then t.rebill_number end) as avg_length_rebill_cycle, \
round(count(distinct case when rebill_number=1 and order_status="redunded" then t.subscription_id end)\
*1.0/count(distinct case when rebill_number=1 then t.subscription_id end), 2) as "%_refunded_customers",\
round(count(distinct case when rebill_number=1 and dispute_date !="nan" then t.subscription_id end)\
*1.0/count(distinct case when rebill_number=1 then t.subscription_id end), 2) as "%_dispute_customers",\
round(count(distinct case when rebill_number=1 and t.cancel_date !="nan" then t.subscription_id end)\
*1.0/count(distinct case when rebill_number=1 then t.subscription_id end), 2) as "%_cancel_customers"\
from purchases as t group by 1', con)
result
```

```python
# + розраховуємо ARPS
#Expected ARPS = (((new_subs - (new_subs * refund_rate)) * trial_price+
#(new_subs-(new_subs*refund_rate)) * CR to purchase * rebill_price * rebill_rate))*0,95) / new_subs 
result = pd.read_sql_query( 
'select product_name,product_price,"%_subscription_customers", avg_length_rebill_cycle,"%_refunded_customers",\
        "%_dispute_customers","%_cancel_customers",\
        (((all_customers-(all_customers*"%_refunded_customers"))*trial_price\
        +(all_customers-(all_customers*"%_refunded_customers"))*"%_subscription_customers"*\
        product_price*avg_length_rebill_cycle)*0.95)/all_customers as ARPS from\
(select t.product_name,p.trial_price, p.product_price, count(distinct t.subscription_id) as all_customers,\
round(count(distinct case when rebill_number=1 then t.subscription_id end)\
*1.0/count(distinct t.subscription_id), 2) as "%_subscription_customers",\
avg(case when t.rebill_number>=1 and ((julianday(t.order_date)-julianday(t.acquisition_date))/30)<=6 \
then t.rebill_number end) as avg_length_rebill_cycle, \
round(count(distinct case when rebill_number=1 and order_status="redunded" then t.subscription_id end)\
*1.0/count(distinct case when rebill_number=1 then t.subscription_id end), 2) as "%_refunded_customers",\
round(count(distinct case when rebill_number=1 and dispute_date !="nan" then t.subscription_id end)\
*1.0/count(distinct case when rebill_number=1 then t.subscription_id end), 2) as "%_dispute_customers",\
round(count(distinct case when rebill_number=1 and t.cancel_date !="nan" then t.subscription_id end)\
*1.0/count(distinct case when rebill_number=1 then t.subscription_id end), 2) as "%_cancel_customers"\
from purchases as t join products as p on t.product_name=p.product_name group by 1,2,3)as d', con)
result
```
